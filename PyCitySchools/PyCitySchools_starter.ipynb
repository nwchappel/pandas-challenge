```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Show table
school_data_complete
```

```python
# Inspect column data types
school_data_complete.dtypes
```

```python
# Clean up columns by renaming
school_data_complete = school_data_complete.rename(columns={"student_name":"Student Name","gender":"Gender","grade":"Grade","school_name":"School Name","reading_score":"Reading Score","math_score":"Math Score","type":"Type","size":"Size","budget":"Budget"})
school_data_complete
```

```python
# Calculate total Schools
total_schools = school_data_complete["School Name"].nunique()

# Calculate total students
total_students = school_data_complete["Student Name"].count()

# Calculate total budget
total_budget_list = school_data_complete.Budget.unique()
total_budget = total_budget_list.sum()

# Calculate the average math score
total_math_score = school_data_complete["Math Score"].sum()
average_math_score = (total_math_score) / (total_students)
average_math_score = (round(average_math_score, 6))

# Calculate average reading score
total_reading_score = school_data_complete["Reading Score"].sum()
average_reading_score = (total_reading_score) / (total_students)
average_reading_score = (round(average_reading_score, 5))

# Calculate the overall passing rate
overall_passing_rate = ((average_math_score) + (average_reading_score)) / 2
overall_passing_rate = (round(overall_passing_rate, 6))

# Calculate % passing math (70 or greater)
passing_math = school_data_complete.loc[school_data_complete["Math Score"] >= 70, "Math Score"]
passing_math = passing_math.count()
percent_passing_math = (passing_math) / (total_students)
percent_passing_math = (round(percent_passing_math, 8))

# Calculate % passing reading (70 or greater)
passing_reading = school_data_complete.loc[school_data_complete["Reading Score"] >= 70, "Reading Score"]
passing_reading = passing_reading.count()
percent_passing_reading = (passing_reading) / (total_students)
percent_passing_reading = (round(percent_passing_reading, 8))

# Add new columns to existing dataframe
school_data_complete["Total Schools"] = total_schools
school_data_complete["Total Students"] = total_students
school_data_complete["Total Budget"] = total_budget
school_data_complete["Average Math Score"] = average_math_score
school_data_complete["Average Reading Score"] = average_reading_score
school_data_complete["% Passing Math"] = percent_passing_math
school_data_complete["% Passing Reading"] = percent_passing_reading
school_data_complete["Overall Passing Rate"] = overall_passing_rate
```

```python
# List the 15 schools
list_total_schools = school_data_complete["School Name"].unique()

# Look at schools' math and reading scores
schools_scores = school_data_complete[["School Name","Reading Score","Math Score"]]

# Group by school and take average of math and reading scores
schools_grouped = schools_scores.groupby(["School Name"])
average_school_scores = schools_grouped.mean()
average_school_scores
```

```python
# Take Remaining Columns for new table to merge with scores
schools_info = school_data_complete[["School Name","Type","Size","Budget"]]

# Group by school again
school_info_grouped = schools_info.groupby(["School Name"])
school_tsb = school_info_grouped.first()
school_tsb
```

```python
# Merge school_tsb and average_school_scores together

merge_table = pd.merge(school_tsb, average_school_scores, on="School Name")
merge_table
```

```python
# Calculate PER STUDENT BUDGET
per_student_budget = (merge_table["Budget"]) / (merge_table["Size"])
```

```python
# Calculate % PASSING MATH, % PASSING READING, OVERALL PASSING RATE per school

# Calculate total students per school
students_per_school = school_data_complete["School Name"].value_counts()

# Calculate how many students have above 70 in math
number_passing_math = school_data_complete[(school_data_complete["Math Score"] >= 70)]

# Calculate how many students per school have above 70 in math
indv_school_passing_math = number_passing_math.groupby(["School Name"]).count()["Size"]
indv_percent_passing_math = (indv_school_passing_math) / (students_per_school)
indv_percent_passing_math = indv_percent_passing_math * 100

# Calculate how many students have above 70 in reading
number_passing_reading = school_data_complete[(school_data_complete["Reading Score"] >= 70)]

# Calculate how many students per school have above 70 in math
indv_school_passing_reading = number_passing_reading.groupby(["School Name"]).count()["Size"]
indv_percent_passing_reading = (indv_school_passing_reading) / (students_per_school)
indv_percent_passing_reading = indv_percent_passing_reading * 100

# Calculate the overall passing rate per school
overall_passing_rate = ((indv_percent_passing_math) + (indv_percent_passing_reading)) / 2
```

```python
school_list = ["Bailey High School","Cabrera High School","Figueroa High School","Ford High School","Griffin High School","Hernandez High School","Holden High School","Huang High School","Johnson High School","Pena High School","Rodriguez High School","Shelton High School","Thomas High School","Wilson High School","Wright High School"]
```

```python
# Make new data into dataframe

psb = per_student_budget
math = indv_percent_passing_math
reading = indv_percent_passing_reading
overall = overall_passing_rate

school_passing_scores = list(zip(school_list, psb, math, reading, overall)) 

school_passing_scores_df = pd.DataFrame(school_passing_scores, columns = ["School Name","Per Student Budget ($)","% Passing Math", "% Passing Reading", "% Overall Passing Rate"])
school_passing_scores_df
```

```python
# Merge school_passing_scores_df and merge_table
# Set School Name as the index
# Format columns

school_summary_data = pd.merge(merge_table, school_passing_scores_df, on="School Name")
school_summary_data = school_summary_data.set_index("School Name")
school_summary_data["Budget"] = school_summary_data["Budget"].map("${:,.2f}".format)
school_summary_data["Per Student Budget ($)"] = school_summary_data["Per Student Budget ($)"].map("{:,.2f}".format)
school_summary_data = school_summary_data.rename(columns={"Type":"School Type","Size":"Total Students","Budget":"Total School Budget","Math Score":"Average Math Score","Reading Score":"Average Reading Score"})
school_summary_data = school_summary_data[["School Type", "Total Students", "Total School Budget", "Per Student Budget ($)", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
school_summary_data
```
